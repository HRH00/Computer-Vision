```python
import cv2 as cv2
import os
import time

MAX_IMAGES=100
image_counter=0
camera = cv2.VideoCapture(1,cv2.CAP_DSHOW)
PERSON=str(input("Enter your Name:\t"))
dir="../Datastore/Face/"
FilePath=dir+PERSON+"/"

# Check if the "PERSON" directory already exists
if not os.path.exists(FilePath):
    # Create the "PERSON" directory
    os.makedirs(FilePath)
    print("Created directory:", FilePath)
else:
    print("Directory already exists:", FilePath)

while True:
    success, frame = camera.read()
    if success:
        cv2.imshow("Saving Images", frame) # This will open an independent window
        path = FilePath + str(image_counter) + ".jpg"
        cv2.imwrite(path, frame)
        print("Image saved:", path)
        image_counter+=1
        time.sleep(0.1)
    if image_counter>MAX_IMAGES-1:
        camera.release()
        break
    if cv2.waitKey(1) & 0xFF==ord('q'): # quit when 'q' is pressed
        camera.release()
        break
    
    
cv2.destroyAllWindows()
```
